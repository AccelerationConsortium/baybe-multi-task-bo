```python
import numpy as np
import matplotlib.pyplot as plt

round = 24
def branin(x1, x2):
    y = float(
        (x2 - 5.1 / (4 * np.pi**2) * x1**2 + 5.0 / np.pi * x1 - 6.0) ** 2
        + 10 * (1 - 1.0 / (8 * np.pi)) * np.cos(x1)
        + 10
    )
    return y
```

```python
# 10 random seeds and 10 initial points
random_seed_list = [17, 23, 42, 87, 99, 131, 518, 1047, 1598, 2024]
init_pt_list = [[7., 6.], 
                [2., 0.], 
                [9., 14.], 
                [2., 8.], 
                [10., 0.], 
                [10., 15.], 
                [5., 2.], 
                [5., 13.], 
                [7., 13.], 
                [7., 2.]]
```

```python
obj = []

for i in range(len(random_seed_list)):
    # create an empty arrat
    exp_i = []
    # loop over random seeds & initial data
    np.random.seed(random_seed_list[i])
    x1 = init_pt_list[i][0]
    x2 = init_pt_list[i][1]
    y = branin(x1, x2)
    exp_i.append(y)

    # random search for 24 rounds
    for j in range(round):
        x1 = np.random.uniform(-5, 10)
        x2 = np.random.uniform(0, 15)
        y = branin(x1, x2)
        exp_i.append(y)
    obj.append(np.array(exp_i))
```
