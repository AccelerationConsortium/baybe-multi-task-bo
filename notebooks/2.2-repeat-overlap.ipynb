```python
# load data 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ax_data = np.loadtxt('fix-Ax.csv')
baybe_data = np.loadtxt('fix-baybe.csv')
random_data = np.loadtxt('fix-random.csv')   # place holder, plotted as white line below
```

```python
data = [ax_data, baybe_data, random_data]
means = {}
stds = {}
for i, d in enumerate(data):
    x = d.copy()
    best_obj = np.array([np.minimum.accumulate(obj_i) for obj_i in x])
    means[i] = best_obj.mean(axis=0)
    stds[i] = best_obj.std(axis=0) / np.sqrt(best_obj.shape[0])

fig, ax = plt.subplots(figsize=(15, 6))
# ax = fig.add_subplot(111)
x = np.arange(1, len(means[0]) + 1)
colors = ['dodgerblue', 'green', 'red']
algorithm = ['Ax', 'Baybe', 'Random']
for i, algo in enumerate(algorithm):
    ax.plot(x, means[i], label=algo, color=colors[i])
    ax.fill_between(x, means[i] - stds[i], means[i] + stds[i], color=colors[i], alpha=0.2)
ax.axhline(y=0.397887, c='r', ls='--', label = 'True min')  # Actual optimum for this problem
ax.set_xlabel('# of iteration')
ax.set_ylabel('Best observed value')
ax.set_title('Branin Optimization, 10 repeats')
ax.legend()
# ax.set_ylim(-1, 80)
```
